```python
import pycaret
```

```python
from pycaret.datasets import get_data
index = get_data('index')
# shows the datasets it has for practice
```

```python
data = get_data('juice')
```

```python
data.shape
```

```python
from pycaret.classification import *
clf1 = setup(data, target = 'Purchase', session_id=123, log_experiment=True, experiment_name='juice1', normalize = True, feature_selection = True)
# setting up a new experiment
# this tells which of the variables are numeric, categoric, binary etc. If satisfied by how it classified, press enter in the box below
# then it shows several parameters, for eg, here train test ratio 70:30; z square normalization has been done; if there is imbalance, fix with SMOTE
# our target is to find which juice will a customer purchase CH or MM
```

```python
training_data = get_config(variable = "X_train")
training_data
```

```python
training_data.shape
```

```python
models()
# gives a list of all the models that are relevant for this dataset
```

```python
best_model = compare_models()
# compares the models in the list wrt accuracy, recall, precision etc
# accordin to this logistic regression will be the best model
```

```python
best_model
```

```python
rf = create_model('rf', fold = 5)
```

```python
rf
```

```python
tuned_rf = tune_model(rf)
```

```python
tuned_rf
```

```python
plot_model(tuned_rf)
```

```python
plot_model(tuned_rf, plot = 'confusion_matrix')
```

```python
plot_model(tuned_rf, plot = 'feature')
```

```python
interpret_model(tuned_rf)
# feature of shap
# left side is class 0 (CH) and right side is class 1 (MM)
# if loyalCH value is high, there is strong chance customer will purchase class 0 (CH)
# if loyalCH value is low, customer will most probably buy MM
# if sale price of MM is high, customer will purchase CH (class 0)
# CH and MM are two juices
```
