```python
import os
import json
from urllib.parse import urlencode
import requests
#to fetch html code of a webpage
from bs4 import BeautifulSoup as bs
#this goes through line by line (parses) of the web page
import numpy as np 
import pandas as pd
# for data manipulation
import re
# regular expression operators
from datetime import datetime
```

```python
r = requests.get("https://www.imdb.com/chart/top/")

soup = bs(r.content)
# converts to beautiful soup object

contents = soup.prettify()
print(contents)
```

```python
movie_title = []
movie_year = []
movie_rating = []
```

```python
imdb_table = soup.find(class_= "chart full-width")
# found by inspect
```

```python
movie_titlecolumn = imdb_table.find_all(class_="titleColumn")
movie_titlecolumn
```

```python
movie_yearcolumn = imdb_table.find_all(class_="secondaryInfo")
movie_yearcolumn

# we don't need this because the title column includes both title and year since it is a higher tag
```

```python
movie_ratingcolumn = imdb_table.find_all(class_="ratingColumn imdbRating")
movie_ratingcolumn
```

```python
for row in movie_titlecolumn:
  title = row.a.text # extract text from 'a' tag in each row
  movie_title.append(title)
movie_title
```

```python
for row in movie_titlecolumn:
  year = row.span.text # extract text from 'a' tag in each row
  movie_year.append(year)
movie_year
```

```python
for row in movie_ratingcolumn:
  rating = row.strong.text
  movie_rating.append(rating)
movie_rating
```

```python
zipped = zip(movie_title, movie_year, movie_rating)
zipped
```

```python
df = pd.DataFrame(list(zipped), columns = ['Movie Title', 'Year of Release', 'IMDb Rating'])
display(df)
```

```python
movie_df = pd.DataFrame({'Movie Title' : movie_title, 'Year of Release': movie_year, 'IMDb Rating': movie_rating})
# uses pandas
movie_df.head(30)
```
